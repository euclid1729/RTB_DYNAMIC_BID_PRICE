```python
import numpy as np
import matplotlib.pylab as plt 
import random
import pandas as pd
from scipy.stats import beta as Beta, norm
from pandas import DataFrame
class BetaBandit(object):
    def __init__(self, num_options):
        self.trials = zeros(shape=(num_options,), dtype=int)
        self.successes = zeros(shape=(num_options,), dtype=int)
        self.alpha=zeros(shape=(num_options,), dtype=int)
        self.beta = zeros(shape=(num_options,), dtype=int)
        self.num_options = num_options
        self.prior = (1.0,1.0)
        
    def set_target_win_rate(self,target_win_rate):
        self.target_win_rate = target_win_rate
 
    def add_result(self, trial_id, success):
        self.trials[trial_id] = self.trials[trial_id] + 1
        if (success):
            self.successes[trial_id] = self.successes[trial_id] + 1
            self.alpha[trial_id]= self.alpha[trial_id]+1
        else:
            self.beta[trial_id]= self.beta[trial_id]+1
 
    def get_recommendation(self):
        ''' returns selected bandit's index '''
        sampled_theta = []
        for i in range(self.num_options):
            #Construct beta distribution for posterior
            dist = Beta(self.prior[0]+self.alpha[i],
                        self.prior[1]+self.beta[i])
            #Draw sample from beta distribution
            sampled_theta += [norm.pdf(self.target_win_rate - dist.rvs(),loc=0,scale=0.1)]
        # Return the index of the sample with the largest value
        return sampled_theta.index( max(sampled_theta) )

class RTB:
    def __init__(self,mu,sigma,p):
        self.mu = mu
        self.sigma = sigma
        self.p = p
        
    def generate_bid(self):
        if random.random()<self.p:
          dist = norm(loc=self.mu,scale=self.sigma)
          return dist.rvs()
        else:
          return 0

def is_conversion(arm_index,list_rtbs):
    bid_price=[]
    for rtb in list_rtbs:
        bid_price.append(rtb.generate_bid())
    if bandits[arm_index] > max(bid_price):
        return True
    else:
        return False

N = 1000
CPM = 5
target_win_rate = .40
rtb1_bid_probability = 0.7
rtb1_mu = 3
rtb1_std = 0.1
rtb2_bid_probability = 0.3
rtb2_mu = 5
rtb2_std = 0.1
rtb1 = RTB(rtb1_mu,rtb1_std,rtb1_bid_probability)
rtb2= RTB(rtb2_mu,rtb2_std,rtb2_bid_probability)
rtb3=RTB(4,0.1,.5)
list_rtbs=[rtb1,rtb2,rtb3]
bandits=np.arange(1,CPM,0.5)
trials = zeros(shape=(N,bandits.shape[0]))
successes = zeros(shape=(N,bandits.shape[0]))
bb = BetaBandit(bandits.shape[0])
bb.set_target_win_rate(target_win_rate)
for i in range(N):
    choice = bb.get_recommendation()
    conv = is_conversion(choice,list_rtbs)
    bb.add_result(choice, conv)
    trials[i] = bb.trials
    successes[i] = bb.successes
    if i==300:
        #RTB4 enters the bidding env
        rtb4=RTB(3.6,0.01,.9)
        list_rtbs.append(rtb4)
    #if (i+1)%400==0:
        #infalting variance of all bandits .. entering explore phase
    for bandit_id in range(bb.num_options):
        if choice==bandit_id and (bb.trials[bandit_id]+1) % 50==0:
          #print i,bb.trials[bandit_id]+1  
          #print "inflating varianceof bandit %d"%bandit_id
          old_alpha = bb.alpha[bandit_id]
          old_beta = bb.beta[bandit_id]
          #print "old params %s,%s"%(str(old_alpha),str(old_beta))
          new_alpha,new_beta = inflate_beta_variance(old_alpha,old_beta)
          bb.alpha[bandit_id]=new_alpha
          bb.beta[bandit_id]=new_beta
          #print "new params %s,%s"%(str(new_alpha),str(new_beta))
```

```python
#infalting variance 
def inflate_beta_variance(alpha,beta):
    #print alpha,beta
    if alpha ==0 or beta == 0 :
      return (alpha+1,beta+1)
    mu =float(alpha)/(alpha+beta)
    variance = (float(alpha)*beta)/((alpha+beta)**2 + float(alpha+beta+1))
    inflated_variance = variance * 1.20
    new_alpha = -1*float(mu) * (inflated_variance **2 + mu **2 -mu)/ float(inflated_variance) **2 
    new_beta = (inflated_variance **2 + mu **2 -mu) * (mu-1) / float(inflated_variance) **2
    #print new_alpha,new_beta
    return (new_alpha,new_beta)
```

```python
def plot_beta(x):
    plt.plot(np.arange(0,1,.05),[beta(x,x).pdf(i) for i in np.arange(0,1,.05)])
    plt.title("BETA PLOT alpha=%f beta =%f"%(x,x))
```
